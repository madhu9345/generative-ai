```python
import cv2
import os

def create_video_from_three_images(image_paths, output_video_path, frame_rate=1, video_size=(640, 480)):
    # Define the codec and create a VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for .mp4
    out = cv2.VideoWriter(output_video_path, fourcc, frame_rate, video_size)

    for image_path in image_paths:
        img = cv2.imread(image_path)

        # Resize image to the video size
        img = cv2.resize(img, video_size)

        # Write frame
        out.write(img)

    out.release()

# List of image paths (only the first three used in the tour)
image_paths = [
    "/content/base.jpg",
    "/content/panromic.jpg",
    "/content/side.png"
]

# Output video path
output_video_path = "/content/eiffel_tour_short.mp4"

# Create the video
create_video_from_three_images(image_paths, output_video_path)
print("Short Eiffel Tower video created successfully!")
```

```python
# Install required libraries
# !pip install gTTS moviepy

from gtts import gTTS
from moviepy.editor import *

# Narration text for the tour
narrations = [
    "Let's start the tour of the Eiffel Tower!",
    "This is the base of the Eiffel Tower. It's made of iron and was constructed in 1887.",
    "As we go up, we can see the panoramic views of Paris.",
    "At this level, we can view landmarks such as the Louvre and the Arc de Triomphe.",
    "That concludes our brief virtual tour. Thank you for joining!"
]

# Corresponding images (place these in your working directory)
image_files = ["base.jpg", "panromic.jpg", "side.png"]

# Step 1: Generate voiceover audio files using gTTS
audio_files = []
for i, text in enumerate(narrations):
    tts = gTTS(text)
    filename = f"voiceover_{i}.mp3"
    tts.save(filename)
    audio_files.append(filename)

# Step 2: Create video clips from images and audio
clips = []

# For the first 3 narrations that have images
for i in range(3):
    audio = AudioFileClip(audio_files[i])
    image = ImageClip(image_files[i]).set_duration(audio.duration).set_audio(audio).resize(height=720)
    clips.append(image)

# For the 4th narration, no new image, just voice over the last image
audio4 = AudioFileClip(audio_files[3])
last_image = ImageClip(image_files[-1]).set_duration(audio4.duration).set_audio(audio4).resize(height=720)
clips.append(last_image)

# For the 5th narration (closing), fade to black with audio
audio5 = AudioFileClip(audio_files[4])
black_clip = ColorClip(size=(1280, 720), color=(0, 0, 0), duration=audio5.duration).set_audio(audio5)
clips.append(black_clip)

# Step 3: Concatenate all clips
final_video = concatenate_videoclips(clips, method="compose")

# Step 4: Export the final video
final_video.write_videofile("eiffel_tower_virtual_tour.mp4", fps=24)
```
