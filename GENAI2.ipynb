**sentimental analysis**

```python
from transformers import pipeline
import gradio as gr

# Load a pre-trained sentiment analysis model
sentiment_analyzer = pipeline("sentiment-analysis")

# Function to analyze sentiment
def analyze_sentiment(text):
    result = sentiment_analyzer(text)[0]
    return {result['label']: result['score']}

# Create a Gradio interface
demo = gr.Interface(
    fn=analyze_sentiment,
    inputs=gr.Textbox(lines=3, placeholder="Enter a sentence..."),
    outputs=gr.Label(),
    title="Sentiment Analysis",
    description="Enter text to determine if it's positive or negative."
)

# Run the app
if __name__ == "__main__":
    demo.launch()
```

**text classification**

```python
# Import necessary libraries
import gradio as gr
from transformers import pipeline

# Load a pre-trained text classification model from Hugging Face
classifier = pipeline("text-classification", model="SamLowe/roberta-base-go_emotions")

# Define the function for text classification
def classify_text(text):
    results = classifier(text)
    return {result['label']: result['score'] for result in results}

# Create a Gradio interface
description = "Enter a piece of text to classify its emotion or category using a pre-trained model."
app = gr.Interface(
    fn=classify_text,  # Function to call
    inputs="text",     # Input type
    outputs="label",   # Output type (shows labels and scores)
    title="Text Classification App",
    description=description
)

# Launch the app
app.launch()
```

**text summarization**

```python
from transformers import pipeline
import gradio as gr

# Load a pre-trained text summarization model from Hugging Face
summarizer = pipeline("summarization")

# Define the function for text summarization
def summarize_text(text):
    # Generate summary with specified parameters
    summary = summarizer(text, max_length=130, min_length=30, do_sample=False)
    return summary[0]['summary_text']

# Create a Gradio interface
description = "Enter a piece of text to generate a concise summary using a pre-trained model."
app = gr.Interface(
    fn=summarize_text,  # Function to call for summarization
    inputs=gr.Textbox(lines=10, placeholder="Enter text to summarize..."),  # Input type (textbox for longer text)
    outputs=gr.Textbox(label="Summarized Text"),  # Output type (textbox for the summary)
    title="Text Summarization App",
    description=description,
)

# Launch the app
if __name__ == "__main__":
    app.launch()
```

**text translator using helsinki**

```python
from transformers import MarianMTModel, MarianTokenizer

def translate_text(text, target_language):
    # Define the model name based on the target language
    model_name = f"Helsinki-NLP/opus-mt-en-{target_language}"

    # Load the tokenizer and model
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    # Prepare the text for translation
    inputs = tokenizer(text, return_tensors="pt", padding=True)

    # Generate translation
    translated = model.generate(**inputs)

    # Decode the translated text
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)

    return translated_text

if __name__ == "__main__":
    input_text = input("Enter the text you want to translate: ")

    # List of available languages (language codes)
    print("Available languages: fr (French), es (Spanish), de (German), it (Italian), ru (Russian)")
    target_language = input("Enter the target language code: ")

    translated_text = translate_text(input_text, target_language)
    print(f"Translated Text ({target_language}): {translated_text}")
```

**text to audio**

```python
from gtts import gTTS

def text_to_speech(text):
    # Convert text to speech
    tts = gTTS(text=text, lang='en', slow=False)

    # Save the audio file
    tts.save("output.mp3")
    print("Speech has been saved as 'output.mp3'.")

    # Play the audio file
    try:
        import os
        os.system("start output.mp3")  # For Windows
        # os.system("afplay output.mp3")  # For macOS
        # os.system("xdg-open output.mp3")  # For Linux
    except Exception as e:
        print(f"Could not play audio: {e}")

if __name__ == "__main__":
    input_text = input("Enter the text you want to convert to speech: ")
    text_to_speech(input_text)
```
